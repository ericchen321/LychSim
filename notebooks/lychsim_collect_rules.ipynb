# LychSim Procedural Rule Collection

```python
import json
import os

import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from tqdm.notebook import tqdm

from lychsim.api import LychSim
from lychsim.utils import EasyDict

import collect_rules_helpers
```

```python
sim = LychSim(server_name='localhost', port=9000, width=1920, height=1080)
print(sim.get_status())
```

```python
rules = dict(vehicle_traj=[], pedestrian_traj=[], road=[], sidewalk=[])
scene_state = EasyDict(sim=sim, rules=rules)

scene_state.scene_name = "NYC_East_Village"

if os.path.isdir(scene_state.scene_name):
    scene_state.rules['vehicle_traj'] = json.load(open(os.path.join(scene_state.scene_name, 'vehicle_traj.json')))
    scene_state.rules['pedestrian_traj'] = json.load(open(os.path.join(scene_state.scene_name, 'pedestrian_traj.json')))
    scene_state.rules['road'] = json.load(open(os.path.join(scene_state.scene_name, 'road.json')))
    scene_state.rules['sidewalk'] = json.load(open(os.path.join(scene_state.scene_name, 'sidewalk.json')))

for k in scene_state.rules:
    print(f"{k}: {len(scene_state.rules[k])} rules loaded")
```
