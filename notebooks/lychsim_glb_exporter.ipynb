## Export Static Meshes

Run `get_mesh` below to export all static meshes under `/Game/<project>/<Mesh>` (SM-prefixed assets by default) to both glb and gltf, and save them into `Plugins/LychSim/Outputs/Mesh`. You can override the project name, mesh folder, or output path via parameters.

```python
import json

def get_mesh(project_name=None,
             mesh_folder="Mesh",
             output_dir="Plugins/LychSim/Outputs/Mesh",
             max_assets=None,
             recursive=True,
             glb_only=False,
             keep_extras=False):
    """
    Export static meshes under `/Game/<project>/<mesh_folder>` to .glb/.gltf.
    Args:
        project_name: Folder between /Game/ and mesh folder.
        mesh_folder: Folder containing SM assets (Mesh or mesh).
        output_dir: Destination directory.
        max_assets: Limit how many to export (None = all).
        recursive: Whether to search subfolders.
        glb_only: If True, export only .glb.
        keep_extras: If False and gltf is produced, delete bin/png extras.
    """
    mesh_root = f"/Game/{project_name}/{mesh_folder}"
    cmd = f"lych obj export_meshes {mesh_root} -out={output_dir}"
    if max_assets is not None:
        cmd += f" -max={max_assets}"
    if not recursive:
        cmd += " -norecursive"
    if glb_only:
        cmd += " -glb_only"
    if keep_extras:
        cmd += " -keep_extras"
    resp = sim.client.request(cmd)
    try:
        data = json.loads(resp)
        print(f"status={data.get('status')} exported={data.get('exported_count')}")
        return data
    except Exception:
        print(resp)
        return resp
```

```python
get_mesh(
    max_assets=None,  
    recursive=True,    
    glb_only=False,   
    keep_extras=False, 
    output_dir="Plugins/LychSim/Outputs/Mesh",
)
```
